## Exercici 2. De la matriu de l'exercici 1, calcula el valor mitjà dels valors introduïts i resta la mitjana resultant de cada un dels valors de la matriu.

```python
print("Mitja: "+str(np.mean(x)))

items=len(x)
print("Número d'iteracions: " + str(items))
y = np.empty(items)

for i in range(items):
   y[i]=(x[i]-np.mean(x))

print("Matriu diferència valor-mitja: "+str(y))
    
```

## Exercici 3. Crea una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

```python
arr55=np.array([[1,56,23,87,96],[4,5,6,98,7]],dtype=int)
print("Matriu 2-D de 5x5: " + str(arr55))

##Mètode 1 per obtenir el valor absolut
items02=arr55.size #total de valors a la matriu
print("Total valors a la matriu: " + str(items02))

newarr55 = arr55.reshape(1,items02)
print("Matriu convertida a 1D: " + str(newarr55))
print("Valor màxim absolut de la matriu - Mètode 1: " + str(np.amax(arr55)))

##Mètode 2 per obtenir el valor absolut
max_total = arr55.max()  
print("Valor màxim absolut de la matriu - Mètode 2: "+str(max_total))

max_relativo_eje01 = arr55.max(axis=1)
print("Valors màxims relatius de la matriu: " + str(max_relativo_eje01))

```

## Exercici 4. Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
## Exemple 1. Mateixa forma (same shape)
a = np.array([1.0, 2.0, 3.0])
b = np.array([2.0, 2.0, 2.0])
print("a * b: "+str(a * b))

## Exemple 2. Escalar x Matriu
c = np.array([1.0, 2.0, 3.0])
d = 3.0
print("c * d: "+str(c * d))

## Exemple 3. Here the newaxis index operator inserts a new axis into a, 
## making it a two-dimensional 4x1 array. Combining the 4x1 array with b, 
## which has shape (3,), yields a 4x3 array
e = np.array([0.0, 10.0, 20.0, 30.0])
f = np.array([1.0, 2.0, 3.0])
print("e + f: " + str(e[:, np.newaxis] + f))

## Exemple 4. Una matriu cal que tingui 1D
bb=np.array([2.0,5.0])
print("Shape d'a: " + str(a.shape))
aa = a.reshape(3,1)
print("Conversió d'a i Shape d'aa: " + str(aa.shape))
print("Matriu aa: " + str(aa))
print("Matriu bb: " + str(bb))
print("aa * bb: " + str(aa * bb))
print("aa + bb: " + str(aa+bb))
```

## Exercici 6. Mask la matriu anterior, realitzeu un càlcul booleà vectoritzat, agafant cada element i comprovant si es divideix uniformement per quatre. Això retorna una matriu de mask de la mateixa forma amb els resultats elementals del càlcul.

```python
import numpy.ma as ma

if (matriu01.all()/4) > 0:
    matriu03= ma.masked_array(matriu01/4, mask=[0,0,0,0,0,0,0,0])
print(str(matriu03))
```

## Exercici 7. A continuació, utilitzeu aquesta màscara per indexar a la matriu de números original. Això fa que la matriu perdi la seva forma original, reduint-la a una dimensió, però encara obteniu les dades que esteu cercant.

```python
print("Accés a matriu03 - Mètode 1: " + str(matriu03[~matriu03.mask]*4))
print("Accés a matriu03 - Mètode 2: " + str(matriu03.compressed()*4))
```
